```python
#Keeping things in latex format
from sympy.interactive import printing
printing.init_printing(use_latex=True)
from sympy import *
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm


TLOCAL = 288
PLOCAL = 1
RLOCAL = 1.2
```

```python
class atmosphericCond:
    def __init__(this,geoAlt):
        if (geoAlt>=0 and geoAlt<11):
            this.tempK = 288.15
            this.pressAtm = 1
            this.tempGradK_km = -6.50
            this.zone=0
            this.geoLimit = 0
        elif (geoAlt>=11 and geoAlt<20):
            this.tempK = 216.65
            this.pressAtm = 0.22
            this.tempGradK_km = 0
            this.zone = 1
            this.geoLimit = 11
        elif (geoAlt>=20 and geoAlt < 32):
            this.tempK = 216.65
            this.pressAtm = 0.05
            this.tempGradK_km = 1
            this.zone = 2
            this.geoLimit = 20
        elif (geoAlt>=32 and geoAlt < 47):
            this.tempK = 228.65
            this.pressAtm=0.01
            this.tempGradK_km = 2.80
            this.zone=3
            this.geoLimit=32
        elif(geoAlt>=47 and geoAlt < 51):
            this.tempK = 270.65
            this.pressAtm=0
            this.tempGradK_km = 0
            this.zone=4
            this.geoLimit=47
        elif (geoAlt>=51 and geoAlt < 71):
            this.tempK = 270.65
            this.pressAtm=0
            this.tempGradK_km=-2.80
            this.zone=5
            this.geoLimit=51
        elif (geoAlt>=71 and geoAlt < 84.85):
            this.tempK=214.65
            this.pressAtm=0
            this.tempGradK_km=-2
            this.zone=6
            this.geoLimit=71
        elif (geoAlt>=84.85):
            this.tempK=186.95
            this.pressAtm=0
            this.tempGradK_km=0
            this.zone=7
            this.geoLimit=84.85

REARTH=6371
def geoPotentialAlt(alt):
    return (REARTH*alt)/(REARTH+alt)

def TemperatureModel(alt):
    h=geoPotentialAlt(alt)
    zone=atmosphericCond(alt)
    return zone.tempK + (zone.tempGradK_km)*(h-zone.geoLimit)

def PressureModel(alt):
    #pr = p_base*e^(-GMRdh/T_base) if dT=0 (barometric equation)
    #pr = p_base*(T_base/T)^(GMR/dT) if dT is not 0 
    #GMR is a constant and is equal to 34.163195
    h=geoPotentialAlt(alt)
    zone=atmosphericCond(alt)
    T = TemperatureModel(alt)
    if (zone.tempGradK_km == 0):
        delPress = zone.pressAtm*np.e**(-34.163195*(h-zone.geoLimit)/zone.tempK)
    else:
        delPress = zone.pressAtm*(zone.tempK/T)**(34.163195/zone.tempGradK_km)
    return delPress*PLOCAL

def DensityModel(alt):
    #Starting with the ideal gas law: PV=nRT, density is approx n/V = P/RT
    #Ratio of density = p_r(Talt/T_base)
    p = PressureModel(alt)
    p_r = p/PLOCAL
    zone = atmosphericCond(alt)
    T = TemperatureModel(alt)
    return p_r/(T/zone.tempK)
        
```

```python

altitude = np.linspace(0,85,1000)
D = [DensityModel(alt) for alt in altitude]
T = [TemperatureModel(alt) for alt in altitude]
P = [PressureModel(alt) for alt in altitude]

fig,[ax,ax1,ax2] = plt.subplots(3,1,figsize=[10,8])

axes = [ax,ax1,ax2]
colors = cm.rainbow(np.linspace(0, 1, 7))
zones=[11,20,32,47,51,71,84.85]
z=0
for a in axes:
    for zone in zones:
        a.axvline(zone,color=colors[z])
        z=z+1
    z=0
ax.plot(altitude,T)
ax1.plot(altitude,P)
ax2.plot(altitude,D)
plt.show()
```
